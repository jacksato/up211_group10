#### Creating a Bargraph that will compare Renter vs Owned Housing Characteristics

```python
# Define columns for renter-occupied and owner-occupied
renter_columns = [
    'Est_Renter_Occupied_Housing_Built_2020_or_later',
    'Est_Renter_Occupied_Housing_Built_2010-2019',
    'Est_Renter_Occupied_Housing_Built_2000-2009',
    'Est_Renter_Occupied_Housing_1Bd',
    'Est_Renter_Occupied_Housing_2_3Bd',
    'Est_Renter_Occupied_Housing_4Bd',
    'Est_Renter_Occupied_Housing_Nocar',
    'Est_Renter_Occupied_Housing_1car'
]

owner_columns = [
    'Est_Owner_Occupied_Housing_Built_2020_or_later',
    'Est_Owner_Occupied_Housing_Built_2010-2019',
    'Est_Owner_Occupied_Housing_Built_2000-2009',
    'Est_Owner_Occupied_Housing_1Bd',
    'Est_Owner_Occupied_Housing_2_3Bd',
    'Est_Owner_Occupied_Housing_4Bd',
    'Est_Owner_Occupied_Housing_Nocar',
    'Est_Owner_Occupied_Housing_1car'
]
```

```python
# Calculate the sum of each category for both renter-occupied and owner-occupied
renter_counts = [df2[col].sum() for col in renter_columns]
owner_counts = [df2[col].sum() for col in owner_columns]
```

```python
# Bar graph with grouped bars
bar_width = 0.35
bar_positions = range(len(renter_columns))
```

```python
#Plot variables and axis on graph, separating variables by Color

plt.bar(bar_positions, renter_counts, bar_width, color='skyblue', label='Renter')
plt.bar(bar_positions, owner_counts, bar_width, color='lightgreen', label='Owner', bottom=renter_counts)

plt.xlabel('Housing Features')
plt.ylabel('Total Count')
plt.title('Comparison of Renter vs Owner Housing Features')
plt.xticks(bar_positions, [col.split('_')[-1] for col in renter_columns], rotation=45)
plt.legend()
```

```python
#Show bar graph
plt.show()
```

#### Joining Census Tract Data with the Housing Data

```python
#maps, import geopandas
import geopandas as gpd
```

```python
# read in a geojson file 
tracts=gpd.read_file('data/Census_Tracts_2020.geojson')
tracts.head()
```

```python
# Plotting the data previously imported
tracts.plot(figsize=(12,10))
```

```python
# Checking the data, looking at its dtypes, and variables 
tracts.info(verbose=True, show_counts=True)
```

```python
# Subsetting the data to get FIPS and geometry
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
# Creating a FIPS column 
tracts['FIPS'] ='06' + '037' + tracts['CT20']
tracts.head()
```

```python
# Merging the dataframe containg housing characteristics with the dataframe previouly imported.
tracts_housing=tracts.merge(df2,on="FIPS")
```

```python
#Checking to see if the two data frames have been merged
tracts_housing.head()
```

#### Show the Spatial Distribution of Housing (Total vs Renter vs Owned)

#### Creating a Map with Folium

##### I will create a map to show the spatial distribution of housing ownership

```python
#Importing the folium library
import folium
```

```python
# Creating the folium map
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')
```

```python
# Plot chorpleth over base map
folium.Choropleth(
                  geo_data=tracts_housing, # geodata
                  data=tracts_housing, # data          
                  key_on='feature.properties.FIPS', 
                  columns=['FIPS','Est_Owner_Total_Occupied','Est_Renter_Total_Occupied'], 
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, 
                  legend_name='Housing Ownership (2020)').add_to(m)    # name on the legend color bar
```
