## If we want to see this data visually with a map we need to import geopandas and have a geojson file in order to map it. 

```python
import geopandas as gpd
```

```python
tracts=gpd.read_file('unemployed.geojson')
tracts.head()
```

## Next we will make a seperate column named FIPS to correspond with our csv file column 'FIPS'. Important note: because our tract column includes the full extent of tract id we just copy it while making the new column 'FIPS'

```python
tracts['FIPS'] = tracts['tract']
```

```python
tracts.head()
```

## I will now use the geometry column to plot the boundary of our data

```python
tracts.plot(figsize=(12,10))
```

## Now we want to look at what data types we have

```python
tracts.info()
```

```python
tracts = tracts[['FIPS','geometry']]
tracts.head()
```

## Finally we can merge our geojson data with our csv file that we have been working on 

```python
tracts_unemployed=tracts.merge(df2,on="FIPS")
```

```python
tracts_unemployed.head()
```

## Now we can create a map and see where there are high concentration of unemployed African Americans. I decided to use natural breaks as other maps would have shown different and not very logical visual representation of unemployment of African Americans throughout LA County. 

```python
tracts_unemployed.plot(figsize=(12,10),
                 column='Unemployed',
                 legend=True, 
                 scheme='NaturalBreaks')
```

## I will import folium in order to produce an interactive map

```python
import folium
```

## Now we can create a folium map and see an interactive map where unemployment of African Americans is highest among census tracts

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts_unemployed, # geo data
                  data=tracts_unemployed, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS', 'Unemployed'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='African Americans Unemployed (2022)').add_to(m)    # name on the legend color bar
m
```
